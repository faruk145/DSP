```python
import matplotlib.pyplot as plt
import numpy as np
def signall(Fs, N, k):
    #Fs: Sampling frequency
    #N: Sample number
    #k: Period number
    F = 100
    T = 1 / F
    t = np.linspace(-k*T/2, k*T/2, 1000)
    xt = np.sinc(2*F*t)**2
    Ts = 1/Fs
    n = np.arange(-N/2, N/2)
    xn = np.sinc(2*F*n*Ts)**2
    #Fourier transform
    w = np.arange(-np.pi, np.pi, 2*np.pi/N)
    xw = np.fft.fftshift(np.fft.fft(xn,N)/N)
    plt.figure()
    plt.subplot(1,3,1)
    plt.plot(t,xt)
    plt.xlabel("t (sn)")
    plt.title("$x(t)$")
    
    plt.subplot(1,3,2)
    plt.stem(n,xn)
    plt.xlabel("t (sn)")
    plt.title("$x[n]$")
    
    plt.subplot(1,3,3)
    plt.stem(w/np.pi, abs(xw))
    plt.xlabel("$\omega$ / $\pi$")
    plt.title("$X(\omega)$ (Amplitude)")
    plt.show()
signall(1000,30,3)
signall(600,30,5)
signall(300,30,5)
signall(100,30,30)
```

```python
import matplotlib.pyplot as plt
import numpy as np
F = 100
T = 1/F
k = 4
t = np.arange(0., k*T, 0.002/22)
x = 5*np.cos(200*np.pi*t)
plt.figure()
plt.xlabel("t")
plt.title("x(t)=5*cos(200*pi*t)")
plt.plot(t,x)
plt.show()
Fs = 2200
Ts = 1/Fs
N = 22
n = np.arange(0,k*N)
xn = 5*np.cos(200*np.pi*n*Ts)
plt.figure()
plt.stem(n,xn)
plt.title("x[n]")
plt.xlabel("sample")
plt.ylabel("x[n]")
plt.show()

x_sinc = np.zeros([k*N,np.size(x,axis=0)])
plt.figure()
for ni in n:
    x_sinc[ni,:]=xn[ni]*np.sinc((t-(ni)*Ts)/Ts)
    plt.plot(x_sinc[ni, :])
plt.title("In time domain, convolution of the signal and sinc(.)")
plt.xlabel("örnek")
plt.show()

xr_t = np.sum(x_sinc, axis = 0)
plt.figure()
plt.xlabel("t")
plt.title("x_r(t)")
plt.plot(t, xr_t)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
F = 100
T = 1/F
k = 2
t = np.arange(0.,k*T,0.00001)
x = 5*np.cos(200*np.pi*t)

plt.figure()
plt.xlabel("t")
plt.title("x(t)")
plt.plot(t,x)
plt.show()

Fs = 2200
Ts = 1/Fs
N = 22
n = np.arange(0., k*N)
xn = 5*np.cos(200*np.pi*n*Ts)

def quantize(x, S):
    X = x.reshape((-1,1))
    S = S.reshape((1,-1))
    dists = abs(X-S)
    
    nearestIndex = dists.argmin(axis = 1)
    quantized = S.flat[nearestIndex]
    
    return quantized.reshape(x.shape)
k_bit = 2
Max_range = np.amax(xn)-np.amin(xn)
q = pow(2,k_bit)
delta = Max_range/q

S_midrise = -np.amax(xn) + delta/2 + np.arange(q)*delta
S_midtread = -np.amax(xn) + np.arange(q)*delta

y_midtrise = quantize(xn, S_midrise)
y_midtread = quantize(xn, S_midtread)

plt.figure()
fig, ax = plt.subplots()
ax.stem(n,xn)
ax.bar(n,y_midtrise,color='orange',edgecolor="darkorange")
plt.xlabel("sample")
plt.title("Quantized signal with midrise method")

fig, ax = plt.subplots()
ax.stem(n,xn)
ax.bar(n,y_midtread,color='orange',edgecolor="darkorange")
plt.xlabel("sample")
plt.title('Quantized signal with midtread method')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
pi = np.pi
Fs = 2200
Ts = 1/Fs
N = 22 # N/22 = k -> when k = 1  N = 22 samples
n = np.arange(0, 2*N, 1)
xn_1 = 7*np.cos(200*pi*n*Ts)
xn_2 = 3*np.cos(200*pi*n*Ts) + 4*np.cos(1800*pi*n*Ts) #pi/11 and 9pi/11 frequency

plt.figure()
plt.stem(n,xn_1)
plt.xlabel("n")
plt.title("$x1[n] = 7cos(\pi n/11)$")
plt.figure()
plt.stem(n,xn_2)
plt.xlabel("n")
plt.title("$x2[n] = 3cos(\pi n/11) + 4cos(9\pi n/11)$")
plt.show()
```

$x1[n] = 7cos(\frac{\pi n}{11})$


$x2[n] = 3cos(\frac{\pi n}{11}) + 4cos(\frac{9\pi n}{11})$

```python
w = np.linspace(-pi,pi-(2*pi/N), N)
x1_w = np.fft.fft(xn_1,N)/N #Fourier transform of x1[n]
x1_w = np.fft.fftshift(x1_w) # We will display X1(w) signal between -pi and +pi

plt.figure()
plt.stem(w/pi, np.abs(x1_w))
plt.title("X1(w)'s Amplitude Graph")
plt.xlabel("w(x$\pi$)")
plt.show()
plt.figure()
plt.stem(w/pi,np.angle(x1_w, deg = True))
plt.title("X1(w)'s Phase Graph")
plt.xlabel("w(x$\pi$)")
plt.show()
```

```python
x2_w = np.fft.fft(xn_2,N)/N #Fourier transform of x2[n]
x2_w = np.fft.fftshift(x2_w) # We will display X2(w) signal between -pi and +pi


plt.figure()
plt.stem(w/pi, np.abs(x2_w))
plt.title("X2(w)'s Amplitude Graph")
plt.xlabel("w(x$\pi$)")
plt.show()
plt.figure()
plt.stem(w/pi,np.angle(x2_w, deg=True))
plt.title("X2(w)'s Phase Graph")
plt.xlabel("w(x$\pi$)")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np 
xn = np.array([1.25 , 3.5 , 4 , 6.7 , -1.4 , -3.5 , -10.5 , 8.5 , 3.8])

def quantize(x, S):
    X = x.reshape((-1,1))
    S = S.reshape((1,-1))
    dists = abs(X-S)
    nearestIndex = dists.argmin(axis=1)
    quantized = S.flat[nearestIndex]
 
    return quantized.reshape(x.shape)

k_bit= 3
Max_range= np.amax(xn)-np.amin(xn) 
q=pow(2,k_bit) 
delta = Max_range/q 
S_midrise = -np.amax(xn) + delta/2 + np.arange(q)*delta # mid-rise formula
S_midtread = -np.amax(xn) + np.arange(q)*delta 

y_midtrise = quantize(xn, S_midrise) #Quantization with midrise
y_midtread = quantize(xn, S_midtread) 

n=np.arange(0.,9)

fig, ax = plt.subplots()
ax.stem(n,xn)
ax.bar(n,y_midtrise,color='orange',edgecolor = "darkorange")
plt.xlabel("sample")
plt.title('Quantization with midrise')
fig, ax = plt.subplots() 
ax.stem(n,xn)
ax.bar(n,y_midtread,color='orange',edgecolor = "darkorange") 
plt.xlabel("sample")
plt.title('Quantization with midtread')
plt.show() 
```
