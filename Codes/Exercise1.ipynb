```python
a = 3 -4j
b = 1 + 2j
C = a + b
D = a/b
#Real parts
print(C.real)
print(D.real)
#Imaginary parts
print(C.imag)
print(D.imag)
#Magnitude
print(np.abs(C))
print(np.abs(D))
#Phase
print(np.angle(C))
print(np.angle(D))
```

```python
import matplotlib.pyplot as plt
%matplotlib notebook
n = np.arange(0, 14+0.001, 1) #w0 = 2Ï€/7 -> f = 1/7 -> T = 7
pi = np.pi
xn = 2*np.sin((2*pi*n/7) - (pi/5))
plt.stem(n,xn)
plt.title('Period: N = 7')
plt.xlabel('n')
plt.ylabel('x[n]')
plt.show()
```

```python
#Graph1
n = np.arange(0, 16, 1)
xn1 = np.zeros(16);
for i in range(len(n)):
    xn1[i] = np.exp(pi/5)

plt.figure()
plt.stem(n, xn1)
plt.title('x1[n]')
plt.xlabel('n')
plt.show()

#Graph2
n = np.arange(0, 16, 1)
xn2 = np.zeros(16, dtype = complex);
for i in range(len(n)):
    xn2[i] = np.exp(1j*pi/5)
    
plt.figure(figsize = (3,3))
plt.stem(n, xn2.real ,label = 'real part')
plt.stem(n, xn2.imag, 'r', markerfmt="D" ,label = 'imaginary part')
plt.legend(loc = 'upper right')
plt.title('x2[n]')
plt.xlabel('n')
plt.show()

#Graph3
n = np.arange(0, 16, 1)
xn3 = np.exp(1j*pi*n/5)
plt.figure(figsize = (8,5))
plt.subplot(2,1,1)
plt.stem(n, xn3.real)
plt.title('x3[n] real part')
plt.xlabel('n')
plt.subplot(2,1,2)
plt.stem(n, xn3.imag)
plt.title('x3[n] imaginary part')
plt.xlabel('n')
plt.tight_layout(pad=2.0)

#Graph4
n = np.arange(0, 16, 1)
xn4 = np.zeros(16, dtype = complex);
for i in range(len(n)):
    xn4[i] = np.cos(pi/5) + 1j* np.sin(pi/5)
plt.figure(figsize = (8,5))
plt.subplot(2,1,1)
plt.stem(n, xn4.real)
plt.title('x4[n] real part')
plt.xlabel('n')
plt.subplot(2,1,2)
plt.stem(n, xn4.imag)
plt.title('x4[n] imaginary part')
plt.xlabel('n')
plt.tight_layout(pad=2.0)

#Graph5
n = np.arange(0, 16, 1)
xn5 = np.cos(pi*n/5) + 1j * np.sin(pi*n/5)
plt.figure(figsize = (8,5))
plt.subplot(2,1,1)
plt.stem(n, xn5.real)
plt.title('x5[n] real part')
plt.xlabel('n')
plt.subplot(2,1,2)
plt.stem(n, xn5.imag)
plt.title('x5[n] imaginary part')
plt.xlabel('n')
plt.tight_layout(pad=2.0)

#Graph6
n = np.arange(-10, 5, 1)
xn6 = np.cos(pi*n/5) + 1j * np.sin(pi*n/5)
plt.figure(figsize = (8,5))
plt.subplot(2,1,1)
plt.stem(n, xn6.real)
plt.title('x6[n] real part')
plt.xlabel('n')
plt.subplot(2,1,2)
plt.stem(n, xn6.imag)
plt.title('x6[n] imaginary part')
plt.xlabel('n')
plt.tight_layout(pad=2.0)
```
